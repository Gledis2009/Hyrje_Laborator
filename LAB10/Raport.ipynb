## 2. Eksperimentet e kërkuara

| Eksperimenti | Çfarë studiohet | Madhësia kryesore |
| --- | --- | --- |
| 1 | variacion natyror për të njëjtin kod | histogramë, `T_std`, `CV` |
| 2 | matje të shkurtra kundrejt të gjata | `CV` për `N` të ndryshme |
| 3 | Python kundrejt NumPy | `T_mean`, `CV` |
| 4 | krijimi i të dhënave brenda/jashtë matjes | ndryshimi i `T_mean` |
| 5 | qasje sekuenciale/rastësishme në memorie | koha dhe `CV` |
| 6 | ngarkesa në sfond | rritja e `T_std` dhe `CV` |
| 7 | printimi dhe I/O | rritja e kohës |
| 8 | ekzekutim i gjatë | trendi i kohës në kohë |

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def summarize_times(times):
    times = np.array(times, dtype=float)
    return {
        "T_mean": times.mean(),
        "T_std": times.std(ddof=1),
        "T_min": times.min(),
        "T_max": times.max(),
        "CV": times.std(ddof=1) / times.mean()
    }

def plot_hist(times, title):
    plt.figure(figsize=(7, 4))
    plt.hist(times, bins=20)
    plt.xlabel("T [s]")
    plt.ylabel("Numri i matjeve")
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
# Eksperimenti 1: variacion natyror
M = 100
N = 200_000
times = []

for _ in range(M):
    s = 0
    start = time.perf_counter()
    for i in range(N):
        s += i
    end = time.perf_counter()
    times.append(end - start)

stats = summarize_times(times)
print(stats)
plot_hist(times, "Variacioni natyror i kohës")
```
