## 1. Qëllimi i laboratorit

Qëllimi është të matet fusha magnetike në pika të ndryshme dhe të ndërtohet hartë 2D e intensitetit. Përdoret magnetometri i telefonit për komponentët $B_x$, $B_y$, $B_z$.

Intensiteti total:

$$
B=\sqrt{B_x^2+B_y^2+B_z^2}
$$

## 2. Eksperimentet

- **Eksperimenti A:** matja e fushës magnetike ambientale.
- **Eksperimenti B:** hartëzimi 1D pranë një magneti.
- **Eksperimenti C:** hartëzimi 2D pranë një magneti.
- **Eksperimenti D:** hartëzimi i anomalive magnetike në ambient, opsional.

Të dhënat ruhen si `B_background.csv`, `B_1D.csv` dhe `B_map_2D.csv`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def add_total_field(df):
    df = df.copy()
    df["B_uT"] = np.sqrt(df["Bx_uT"]**2 + df["By_uT"]**2 + df["Bz_uT"]**2)
    return df

df_bg = pd.read_csv("data/B_background.csv")
df_bg = add_total_field(df_bg)

B_background = df_bg["B_uT"].mean()
print("B_background [uT] =", B_background)
print("sigma_background [uT] =", df_bg["B_uT"].std(ddof=1))
```

```python
df1 = pd.read_csv("data/B_1D.csv")
df1 = add_total_field(df1)
df1["DeltaB_uT"] = df1["B_uT"] - B_background

plt.figure(figsize=(7, 4))
plt.plot(df1["x_cm"], df1["DeltaB_uT"], "o-")
plt.xlabel("Largësia x [cm]")
plt.ylabel("Delta B [uT]")
plt.title("Perturbimi magnetik si funksion i largësisë")
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("data/B_map_2D.csv")
df = add_total_field(df)
df["DeltaB_uT"] = df["B_uT"] - B_background

plt.figure(figsize=(6, 5))
plt.scatter(df["x_cm"], df["y_cm"], c=df["B_uT"], s=120)
plt.colorbar(label="B [uT]")
plt.xlabel("x [cm]")
plt.ylabel("y [cm]")
plt.title("Harta 2D e fushës magnetike")
plt.axis("equal")
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 5))
plt.scatter(df["x_cm"], df["y_cm"], c=df["DeltaB_uT"], s=120)
plt.colorbar(label="Delta B [uT]")
plt.xlabel("x [cm]")
plt.ylabel("y [cm]")
plt.title("Harta 2D e perturbimit magnetik")
plt.axis("equal")
plt.grid(True)
plt.show()
```

```python
grid = df.pivot(index="y_cm", columns="x_cm", values="DeltaB_uT")

plt.figure(figsize=(7, 5))
plt.imshow(
    grid.values,
    origin="lower",
    extent=[df["x_cm"].min(), df["x_cm"].max(), df["y_cm"].min(), df["y_cm"].max()],
    aspect="equal"
)
plt.colorbar(label="Delta B [uT]")
plt.xlabel("x [cm]")
plt.ylabel("y [cm]")
plt.title("Heatmap i perturbimit magnetik")
plt.show()
```
