## 1. Qëllimi i laboratorit

Ky laborator studion matjet indirekte. Dendësia nuk matet drejtpërdrejt, por llogaritet nga masa dhe vëllimi:

$$
\rho = \frac{m}{V}
$$

Kërkesa kryesore është të llogaritet edhe pasiguria finale e dendësisë.

## 2. Teoria

Për një paralelepiped:

$$
V = abc
$$

Përhapja e pasigurisë relative:

$$
\frac{\Delta V}{V} \approx \frac{\Delta a}{a}+\frac{\Delta b}{b}+\frac{\Delta c}{c}
$$

Për dendësinë:

$$
\frac{\Delta \rho}{\rho} \approx \frac{\Delta m}{m}+\frac{\Delta V}{V}
$$

```python
import numpy as np

# Zëvendëso me matjet reale.
m_g = np.array([], dtype=float)   # masa në gram
a_cm = np.array([], dtype=float)  # gjatësia
b_cm = np.array([], dtype=float)  # gjerësia
c_cm = np.array([], dtype=float)  # lartësia

# Pasiguritë instrumentale, përshtati sipas instrumenteve.
dm_g = 0.01
da_cm = db_cm = dc_cm = 0.01

if m_g.size and a_cm.size and b_cm.size and c_cm.size:
    m = m_g.mean()
    a = a_cm.mean()
    b = b_cm.mean()
    c = c_cm.mean()

    V = a*b*c
    rho = m/V

    rel_V = da_cm/a + db_cm/b + dc_cm/c
    rel_rho = dm_g/m + rel_V
    drho = rho * rel_rho

    print(f"V = {V:.3f} cm^3")
    print(f"rho = ({rho:.3f} ± {drho:.3f}) g/cm^3")
```

## 5. Përfundim

Dendësia e matur duhet raportuar në formën:

$$
\rho = (\rho_0 \pm \Delta \rho)\,\mathrm{g/cm^3}
$$

Analiza tregon se madhësitë e llogaritura nuk mund të jenë më të sakta se matjet mbi të cilat mbështeten.
