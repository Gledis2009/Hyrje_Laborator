## 1. Qëllimi i laboratorit

Qëllimi është të matet eksperimentalisht koha e ekzekutimit $T$ për algoritme të ndryshme dhe të studiohet varësia nga madhësia e input-it $N$.

Modeli i analizës është:

$$
T(N) \sim N^p
$$

ku $p$ gjendet nga pjerrësia e grafikut log-log.

## 2. Eksperimentet e kërkuara

| Eksperimenti | Algoritmi | Kompleksiteti i pritur |
| --- | --- | --- |
| 1 | një cikël `for` | $N$ |
| 2 | dy cikle të futura | $N^2$ |
| 3 | tre cikle të futura | $N^3$ |
| 4 | `numpy.sort` | $N\log N$ |

Për çdo vlerë të $N$ koha matet disa herë dhe ruhen `T_mean` dhe `T_std` në skedarët `data/eksperimenti1.dat`, ..., `data/eksperimenti4.dat`.

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def time_repeat(func, repeats=10):
    times = []
    for _ in range(repeats):
        start = time.perf_counter()
        func()
        end = time.perf_counter()
        times.append(end - start)
    times = np.array(times)
    return times.mean(), times.std(ddof=1)

def exp_one_loop(N):
    def run():
        s = 0
        for i in range(N):
            s += i
    return run

def exp_two_loops(N):
    def run():
        s = 0
        for i in range(N):
            for j in range(N):
                s += i*j
    return run

def exp_numpy_sort(N):
    a = np.random.rand(N)
    def run():
        np.sort(a)
    return run
```

```python
# Shembull për një eksperiment. Përshtat N_values sipas kohës së kompjuterit.
N_values = np.array([10, 100, 1000, 10000])
rows = []

for N in N_values:
    T_mean, T_std = time_repeat(exp_one_loop(int(N)), repeats=10)
    rows.append((N, T_mean, T_std))

rows = np.array(rows)
print("N, T_mean, T_std")
print(rows)
```

```python
def analyze_timing_table(rows, title, expected):
    N = rows[:, 0]
    T_mean = rows[:, 1]

    plt.figure(figsize=(7, 4))
    plt.loglog(N, T_mean, "o-")
    plt.xlabel("N")
    plt.ylabel("T_mean [s]")
    plt.title(title)
    plt.grid(True, which="both")
    plt.show()

    logN = np.log10(N)
    logT = np.log10(T_mean)
    p, b = np.polyfit(logN, logT, 1)

    print(f"Eksponenti i matur p = {p:.3f}")
    print("Kompleksiteti i pritur:", expected)
    return p

if "rows" in globals() and len(rows) > 1:
    p1 = analyze_timing_table(rows, "Eksperimenti 1: një cikël", "p ≈ 1")
```

## 3. Diskutim

Në grafik log-log, një varësi $T(N)=aN^p$ bëhet vijë e drejtë:

$$
\log_{10}(T)=\log_{10}(a)+p\log_{10}(N)
$$

Devijimet nga vlera teorike vijnë nga overhead-i i Python-it, sistemi operativ, cache, madhësitë shumë të vogla të input-it dhe zhurma e matjes.
