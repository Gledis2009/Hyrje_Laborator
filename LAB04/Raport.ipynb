## 1. Qëllimi i laboratorit

Qëllimi është përdorimi i Python-it në Jupyter për analizë të të dhënave laboratorike. Kërkesat përfshijnë përdorimin e NumPy për llogaritje numerike, Matplotlib për grafikë dhe përshtatje lineare me `numpy.polyfit`.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4], dtype=float)
y = np.array([2.1, 4.2, 6.1, 8.0], dtype=float)

print("x =", x)
print("y =", y)
print("2*x =", 2*x)
```

## 2. Analiza numerike

NumPy përdoret sepse vektorët lejojnë llogaritje të shpejta dhe të qarta. Në vend që të punohet me lista të zakonshme Python, të dhënat eksperimentale vendosen në `numpy.array`.

```python
coeff = np.polyfit(x, y, 1)
slope = coeff[0]
intercept = coeff[1]
fit = slope*x + intercept

print("Pjerrësia =", slope)
print("Prerja me boshtin =", intercept)
```

```python
plt.figure(figsize=(7, 4))
plt.scatter(x, y, label="Të dhënat")
plt.plot(x, fit, label=f"Fit: y = {slope:.2f}x + {intercept:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Shembull i përshtatjes lineare")
plt.grid(True)
plt.legend()
plt.show()
```

## 3. Interpretim

Përshtatja lineare jep një model të formës:

$$
y = ax + b
$$

ku `a` është pjerrësia dhe `b` është prerja me boshtin. Në një raport laboratorik nuk mjafton të shfaqet grafiku; duhet shpjeguar edhe kuptimi fizik/statistikor i parametrave.
