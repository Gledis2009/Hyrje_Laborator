## 1. Qëllimi i laboratorit

Qëllimi është të maten madhësi bazë fizike dhe të raportohet rezultati me pasiguri:

$$
x = x_{\text{mat}} \pm \Delta x
$$

Laboratori kërkon matje të përsëritura për gjatësinë, masën dhe kohën, si dhe diskutim të gabimeve instrumentale, rastësore dhe sistematike.

## 2. Metoda

- Për gjatësinë përdoren vizore, kalibër ose mikrometër.
- Për masën përdoret peshore.
- Për kohën përdoret kronometër ose smartphone.
- Për çdo madhësi kryhen disa matje dhe llogariten mesatarja, devijimi standard dhe pasiguria.

```python
import numpy as np

def summarize(values, resolution=None):
    values = np.array(values, dtype=float)
    mean = values.mean()
    std = values.std(ddof=1) if len(values) > 1 else 0.0
    u_stat = std / np.sqrt(len(values)) if len(values) > 1 else 0.0
    u_instr = resolution / 2 if resolution is not None else 0.0
    u_total = max(u_stat, u_instr)
    return mean, std, u_stat, u_instr, u_total

# Zëvendëso listat me matjet reale.
gjatesia_cm = np.array([], dtype=float)
masa_g = np.array([], dtype=float)
koha_s = np.array([], dtype=float)

if gjatesia_cm.size:
    print("Gjatësia:", summarize(gjatesia_cm, resolution=0.1))
if masa_g.size:
    print("Masa:", summarize(masa_g, resolution=0.01))
if koha_s.size:
    print("Koha:", summarize(koha_s, resolution=0.01))
```
