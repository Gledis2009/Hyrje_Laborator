## 1. Qëllimi i laboratorit

Qëllimi është të tregohet se koha e reagimit është madhësi statistikore. Ajo nuk përshkruhet nga një vlerë e vetme, por nga një shpërndarje.

## 2. Metoda e rënies së vizores

Koha e reagimit llogaritet nga distanca e rënies:

$$
t = \sqrt{\frac{2d}{g}}
$$

ku $d$ është distanca e rënies dhe $g = 9.81\,\mathrm{m/s^2}$.

Kërkohen të paktën 30-50 matje për person.

```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81

# Vendos distancat reale të rënies në centimetra.
d_cm = np.array([], dtype=float)

if d_cm.size:
    d_m = d_cm / 100
    t = np.sqrt(2*d_m/g)

    t_mean = t.mean()
    t_std = t.std(ddof=1)

    print(f"t = ({t_mean:.3f} ± {t_std:.3f}) s")

    plt.figure(figsize=(7, 4))
    plt.hist(t, bins=15, density=True, alpha=0.7, label="Të dhënat")

    x = np.linspace(t.min(), t.max(), 200)
    gaussian = 1/(t_std*np.sqrt(2*np.pi))*np.exp(-(x-t_mean)**2/(2*t_std**2))
    plt.plot(x, gaussian, "r-", label="Gaussian")

    plt.xlabel("Koha e reagimit [s]")
    plt.ylabel("Densitet probabiliteti")
    plt.title("Histogrami i kohës së reagimit")
    plt.grid(True)
    plt.legend()
    plt.show()
```

## 3. Rezultati final

Rezultati raportohet si:

$$
t = (\bar{t} \pm \sigma)
$$

ku $\bar{t}$ është koha mesatare dhe $\sigma$ është devijimi standard i matjeve.

Koha e reagimit ndryshon nga prova në provë për shkak të faktorëve njerëzorë dhe rastësorë. Histogrami tregon formën e shpërndarjes dhe lejon krahasim me modelin Gaussian. Raportimi vetëm i mesatares nuk mjafton.
