## 1. Qëllimi i laboratorit

Qëllimi është të përdoret Markdown në Jupyter Notebook për të shkruar raporte të qarta shkencore. Laboratori kërkon përdorimin e titujve, listave, formatimit të tekstit, formulave LaTeX, figurave dhe links.

## 2. Shembuj të sintaksës Markdown

Titujt krijohen me `#`, `##`, `###`. Teksti mund të jetë **i trashë**, *italic* ose `kod`.

Lista me pika:

- Matjet
- Analiza
- Përfundimi

Lista e numëruar:

1. Përgatitja e eksperimentit
2. Kryerja e matjeve
3. Analiza e të dhënave

## 3. Formula matematikore

Jupyter mbështet LaTeX. Një formulë brenda tekstit shkruhet si $F = ma$.

Formula në rresht të veçantë:

$$
x(t)=x_0+v_0t+\frac{1}{2}at^2
$$

Kjo është e rëndësishme për raportet laboratorike sepse teoria dhe rezultatet duhet të paraqiten qartë.

## 4. Figura dhe links

Figura mund të futet me sintaksën:

```markdown
![përshkrimi](figs/figura.png)
```

Një link i klikueshëm:

```markdown
[Markdown Guide](https://www.markdownguide.org/cheat-sheet/)
```

## 5. Struktura e raportit

Një raport i mirë laboratorik duhet të ketë këtë rrjedhë:

| Seksioni | Përmbajtja |
| --- | --- |
| Qëllimi | çfarë matet ose studiohet |
| Teoria | ligjet dhe formulat |
| Metoda | si u krye eksperimenti |
| Analiza | kod, tabela dhe grafiqe |
| Rezultatet | vlerat numerike dhe figurat |
| Përfundimi | interpretimi dhe kufizimet |

## 6. Përfundim

Markdown e bën notebook-un një raport të lexueshëm, jo vetëm një vend për kod. Përdorimi i titujve, formulave dhe interpretimit është po aq i rëndësishëm sa llogaritjet numerike.
