## 1. Qëllimi i laboratorit

Qëllimi është të përdoret Git për kontroll versionesh dhe GitHub për ruajtje/dorëzim online të raporteve laboratorike. Laboratori kërkon kuptimin e repository-t lokal, repository-t në GitHub, commit-eve dhe workflow-it minimal të dorëzimit.

## 2. Konceptet kryesore

- **Git** ruan historinë e ndryshimeve në projekt.
- **Repository** është dosja e projektit e ndjekur nga Git.
- **GitHub** ruan repository-n online dhe mundëson dorëzimin në GitHub Classroom.
- **Commit** është një version i ruajtur i punës.
- **Push** ngarkon commit-et lokale në GitHub.

## 3. Workflow minimal

Komandat kryesore që përdoren pas përditësimit të një raporti janë:

```bash
git status
git add .
git commit -m "Përditësim i raportit"
git push
```

Para dorëzimit final përdoret një mesazh më i qartë, p.sh.:

```bash
git commit -m "Versioni final i raportit LAB02"
git push
```
