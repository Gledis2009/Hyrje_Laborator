## 2. Eksperimentet

- **Eksperimenti A:** telefoni i palëvizur në qiell të hapur.
- **Eksperimenti B:** telefoni i palëvizur pranë ndërtesës ose brenda.
- **Eksperimenti C:** ecje e ngadaltë përgjatë një rruge të shkurtër.

Të dhënat ruhen në `data/`, p.sh. `gps_outdoor_stationary.csv`, `gps_near_building.csv`, `gps_walk.csv`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load_gps_csv(filename, lat_col="latitude", lon_col="longitude"):
    df = pd.read_csv(filename)
    print(df.columns)
    df = df.dropna(subset=[lat_col, lon_col])

    lat = df[lat_col].to_numpy()
    lon = df[lon_col].to_numpy()

    R = 6371000
    lat_rad = np.deg2rad(lat)
    lon_rad = np.deg2rad(lon)

    lat0 = np.mean(lat_rad)
    lon0 = np.mean(lon_rad)

    x = R * np.cos(lat0) * (lon_rad - lon0)
    y = R * (lat_rad - lat0)
    r = np.sqrt(x**2 + y**2)

    return x, y, r
```

```python
# Përshtat emrat e kolonave sipas CSV-së tënde.
x, y, r = load_gps_csv("data/gps_outdoor_stationary.csv",
                       lat_col="latitude",
                       lon_col="longitude")

print("Numri i pikave:", len(r))
print("sigma_x [m]:", np.std(x, ddof=1))
print("sigma_y [m]:", np.std(y, ddof=1))
print("r_68 [m]:", np.percentile(r, 68))
print("r_95 [m]:", np.percentile(r, 95))
print("r_max [m]:", np.max(r))
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=10)
plt.axhline(0, linestyle="--", linewidth=1)
plt.axvline(0, linestyle="--", linewidth=1)
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.title("Shpërndarja 2D e pozicionit GPS")
plt.axis("equal")
plt.grid(True)
plt.show()

plt.figure(figsize=(7, 4))
plt.hist(r, bins=30)
plt.xlabel("Devijimi radial [m]")
plt.ylabel("Numri i matjeve")
plt.title("Histogrami i gabimit radial")
plt.grid(True)
plt.show()
```
