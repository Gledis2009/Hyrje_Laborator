## 2. Gjenerimi i të dhënave

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data/model.dat", delimiter=",", comments="#")

x_Data = data[:, 0]
y_Data = data[:, 1]

print("Numri i pikave:", len(x_Data))
print("x =", x_Data)
print("y =", y_Data)
````

````python
a, b = np.polyfit(x_Data, y_Data, 1)

print(f"Pjerrësia a = {a:.3f}")
print(f"Prerja me boshtin b = {b:.3f}")
````

````python
xfit = np.linspace(x_Data.min(), x_Data.max(), 100)
yfit = a*xfit + b

plt.figure(figsize=(8, 5))
plt.plot(x_Data, y_Data, "ok", label="Të dhënat")
plt.plot(xfit, yfit, "-r", lw=2, label=f"y = {a:.2f}x + {b:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Përshtatja lineare e të dhënave")
plt.grid(True)
plt.legend()
plt.show()
````

````python
residuals = y_Data - (a*x_Data + b)

plt.figure(figsize=(8, 5))
plt.stem(x_Data, residuals)
plt.axhline(0, ls="--", color="k", lw=1)
plt.xlabel("x")
plt.ylabel("mbetjet")
plt.title("Analiza e mbetjeve")
plt.grid(True)
plt.show()

print("Mesatarja e mbetjeve:", residuals.mean())
print("Devijimi standard i mbetjeve:", residuals.std(ddof=1))
````

## 3. Interpretimi

Pjerrësia `a` tregon sa ndryshon `y` kur `x` rritet me një njësi. Parametri `b` është vlera e modelit kur `x = 0`. Nëse mbetjet shpërndahen rreth zeros pa trend të qartë, modeli linear është i përshtatshëm për të dhënat.
